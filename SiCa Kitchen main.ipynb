```python
import requests
# import openai didn't end up using GPT for this part
import json
```

```python
# List of available ingredients
ingredients = "chicken", "tomatoes", "carots", "eggs", "parmesan", "rice", "peas", "spinach", "sour cream", "beef", "onions", "potatoes"
```

```python
# Importing recipe API
url = 'https://api.spoonacular.com/recipes/findByIngredients'
api_key = "7df272e674dd49fb97eaa1cad6d190f9" 
params = {
    'ingredients': ingredients,
    'number' : 500, # getting 500 recipes using our ingredients
    'apiKey': api_key,
    'cheap': True
}

response = requests.get(url, params= params)
```

```python
# Checking if the API call worked
if response.status_code == 200:
    recipes = response.json() # Storing the gathered recipes in a list of dictionaries
else:
    print("Error", response.status_code)
```

```python
# Simplifying the information for each recipe 
possible_recipes = [] 
for recipe in recipes:
    url = f'https://api.spoonacular.com/recipes/{recipe["id"]}/information'
    params2 = {'apiKey': api_key}
    recipes_response = requests.get(url, params=params2)
    potential_recipes = recipes_response.json()
    
    simplified_recipe = {
        'title': potential_recipes.get('title', 'No title available'),
        'ingredients': [ingredient['original'] for ingredient in potential_recipes.get('extendedIngredients', [])],
        'instructions': potential_recipes.get('instructions', 'No instructions available'),
        'recipe_url': potential_recipes.get('sourceUrl', 'No source URL available')
    }

    # Creating a list containing all potential recipes
    possible_recipes.append(simplified_recipe)
```
